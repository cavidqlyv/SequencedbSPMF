```python
import csv
import numpy as np
from pprint import pprint 
```

```python

filename ="sequences.csv"
```

```python
SequenceDatabase=[]

with open(filename, 'r') as data:
    for lines in csv.reader(data):
      for line in lines:   
        line = line.replace(" ", "")
        linearr=line.split('(')
     
        betterLine = map(lambda x: x.replace(")",""), linearr)
        sequence=list(betterLine)
        sequence.pop(0)
        sequence[len(sequence)-1]=sequence[len(sequence)-1].replace(";0","")
        sequence[len(sequence)-1]=sequence[len(sequence)-1].replace(";1","")
        SequenceDatabase.append(sequence)
```

```python
SequenceDatabase[0]
```

```python
MinMaxDict={}
```

```python
for sequenceRow in SequenceDatabase:
    index=0
    for sequence in sequenceRow:
        index+=1
        TS0,seq0=sequence.split(",")
        try:
            TS1,seq1=SequenceDatabase[0][index].split(",")

            try:
                # check MAX
                if MinMaxDict[(seq0,seq1)][0]>abs(int(TS0)-int(TS1)):
                    # if true change MIN to new min 
                    MinMaxDict[(seq0,seq1)][0]=abs(int(TS0)-int(TS1))
                # check MAX
                elif MinMaxDict[(seq0,seq1)][1]<abs(int(TS0)-int(TS1)):
                    #if true change MAX to new MAX value
                    MinMaxDict[(seq0,seq1)][1]=abs(int(TS0)-int(TS1))

            except: #create new MINMAX
                MinMaxDict[(seq0,seq1)]=[abs(int(TS0)-int(TS1)),abs(int(TS0)-int(TS1))]
        except:
            index=0
```
